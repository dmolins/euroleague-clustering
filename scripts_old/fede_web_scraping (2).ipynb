```python
import bs4 
import time
from selenium import webdriver 

url = 'https://www.basquetcatala.cat/estadistiques/5d98ee5aa4d4273e2a48fe03'

# Creem un bot que va a la pestanya de "Jugadors" per poder tirar el BeautifulSoup sobre aquella pestanya,
# ja que t√© el mateix URL que les pestanyes "Equip", etc.

browser = webdriver.Chrome(executable_path="./chromedriver")
browser.get(url)

element = browser.find_element_by_id("jugadores")
element.click()
time.sleep(2)


html = browser.page_source
soup = bs4.BeautifulSoup(html,'html.parser')
browser.close()
```

```python
containers = soup.findAll('table')
taula_head = containers[1].find_all('tr', limit=2)

headers = []

for el in taula_head[0].findAll('td')[0:7]:
    headers.append(el.getText())

rows = containers[1].findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td', limit = 7)]
            for i in range(len(rows))]

ps = []

for el in player_stats:
    if 'face' in el[0]:
        ps.append(el)
        
for el in ps:
    
    name = el[0]
    name = name.split(' ')
    el[0] = name[1] + ' ' + name[2] + ' ' + name[3] + ' ' + name[4]
    
    tl = el[3]
    tl = tl.split('\n')
    el[3] = tl[1] + ' ' + tl[2]
    
    t2 = el[4]
    t2 = t2.split('\n')
    el[4] = t2[1] + ' ' + t2[2]
    
    
    t3 = el[5]
    t3 = t3.split('\n')
    el[5] = t3[1] + ' ' + t3[2]
    
    

stats = pd.DataFrame(ps, columns = headers)

stats.to_csv('vs_stpep.csv', index = None, header = True)
```
